```python
# 1) Crie duas função simples sem parametros 
def olamundo():
    print("Hello World!")
    
def saudacao():
    print("Olá, como você está?")

olamundo()
saudacao()
```

```python
# 2) Crie uma função com dois paramentros
def soma(a, b):
    return a + b

soma(2, 5)
```

```python
# 3 Faça o uso da lógica if
def maiornumero(a, b):
    if a > b:
        print(a, "é maior que", b)
    if a == b:
        print(a, "é igual a", b)
    if b > a:
        print(b, "é maior que", a)
maiornumero(3, 5)
```

```python
# 4 Faça o uso da lógica if e else
def par(num):
    if num % 2 == 0:
        print(num, "é par")
    else:
        print(num, "é ímpar")
par(2)
par(4)
par(5)
par(7)
```

```python
# 5) Crie uma função usando usando um retorn
def quadrado(num):
    return num ** 2
quadrado(9)
```

```python
# 6) Crie uma função para um calculo de multiplicação
def multiplicar(a, b):
    return a * b
multiplicar(3, 5)
```

```python
# 7) Crie um for com números
for num in range(1, 11):
    print(num)
```

```python
# 8) Crie um for para strings
minhastring = "Olá, mundo!"
for letra in minhastring:
    print(letra)
```

```python
# 9) Crie um for aninhados
for i in range(1, 6):
    for j in range(1, i+1):
        print(j, end="")
    print()
```

```python
# 10) Crie um uma função com um for
def somarlista(minha_lista):
    total = 0
    for num in minha_lista:
        total += num
    return total

somarlista([3, 4, 5])
```

```python
# 11) Faça o uso do break em alguma função
def encontrarnumero(minha_lista, alvo):
    for num in minha_lista:
        if num == alvo:
            print("Encontrado", alvo, "na lista!")
            break
    else:
        print(alvo, "não foi encontrado na lista.")
encontrarnumero([3, 4, 5, 6, 7, 9], 6)
```

```python
# 12) Faça o uso do continue em alguma função
def imprimirnumerosimpares(minha_lista):
    for num in minha_lista:
        if num % 2 == 0:
            continue
        print(num, "é um número ímpar.")

imprimirnumerosimpares([2, 3, 4, 5, 6, 7, 9])
```
