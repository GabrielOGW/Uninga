```python
resposta=int(input("qual colocação vc quer saber?"))

if resposta == 1:
  print("Santos! Gigante!")
elif resposta == 2:
  print("algum time pior que santos que está em segundo")
elif resposta == 3:
  print("outro time pior que santos que está em terceiro")
elif resposta == 4:
  print("algum time pior que santos que está em quarto")
elif resposta == 5:
  print("algum time pior que santos que está em quinto")
elif resposta == 6:
  print("algum time pior que santos que está em sexto")
else:
  print("só temos 6 times")
```

```python
var = 5

if var == 5:
  print("os valores são iguais")

if var != 7:
  print("o valor é diferente de 7")

if var > 2:
  print("o valor é maior que 2")

if var >= 5:
  print("o valor é maior ou igual a 5")

if var < 7:
  print("o valor é menor que 7")

if var <= 5:
  print("o valor é menor ou igual a 5")
```
