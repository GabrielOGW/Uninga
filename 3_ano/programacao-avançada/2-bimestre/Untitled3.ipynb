```python
class Retangulo:
    def __init__(self, comprimento, largura):
        self.comprimento = comprimento
        self.largura = largura
    
    def calcular_area(self):
        return self.comprimento * self.largura
    
    def calcular_perimetro(self):
        return 2 * (self.comprimento + self.largura)

meu_retangulo = Retangulo(5, 3)
area = meu_retangulo.calcular_area()
perimetro = meu_retangulo.calcular_perimetro()

print("Área:", area)
print("Perímetro:", perimetro)
```

```python
class Calculadora:
    def somar(self, a, b):
        return a + b
    
    def subtrair(self, a, b):
        return a - b
    
    def multiplicar(self, a, b):
        return a * b
    
    def dividir(self, a, b):
        if b != 0:
            return a / b
        else:
            print("Erro: divisão por zero!")

calc = Calculadora()
soma = calc.somar(5, 3)
subtracao = calc.subtrair(10, 7)
multiplicacao = calc.multiplicar(4, 6)
divisao = calc.dividir(15, 3)

print("Soma:", soma)
print("Subtração:", subtracao)
print("Multiplicação:", multiplicacao)
print("Divisão:", divisao)
```

```python
class funcaoPrimeiroGrau:
    def __init__(self, coeficienteA, coeficienteB):
      self.coeficienteA = coeficienteA
      self.coeficienteB = coeficienteB

    def calcularValor(self, x):
      return self.coeficienteA * x + self.coeficienteA

funcao = funcaoPrimeiroGrau(2, 3)

x = 5
valor = funcao.calcularValor(x)
print(f"Para x = {x}, o valor da funcao é: {valor}")
```
