1-Crie uma classe diferente das aulas valores numericos.

```python
class Triangulo:
    def __init__(self, lado_a, lado_b, lado_c):
        self.lado_a = lado_a
        self.lado_b = lado_b
        self.lado_c = lado_c

    def calcular_seno(self):
        return self.lado_a / self.lado_c

    def calcular_cosseno(self):
        return self.lado_b / self.lado_c

    def calcular_tangente(self):
        return self.lado_a / self.lado_b

# Exemplo de uso da classe Triangulo
triangulo = Triangulo(6, 4, 5)
seno = triangulo.calcular_seno()
cosseno = triangulo.calcular_cosseno()
tangente = triangulo.calcular_tangente()

print(f"Seno: {seno}")
print(f"Cosseno: {cosseno}")
print(f"Tangente: {tangente}")
```

2-Crie uma classe usando alguma equação matematica ou da fisica.

```python
class Bhaskara:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def calcRaiz1(self):
        delta = self.b**2 - 4*self.a*self.c
        if delta < 0:
            return None
        raiz_1 = (-self.b + (delta)**0.5) / (2*self.a)
        return raiz_1

    def calcRaiz2(self):
        delta = self.b**2 - 4*self.a*self.c
        if delta < 0:
            return None
        raiz_2 = (-self.b - (delta)**0.5) / (2*self.a)
        return raiz_2

bhaskara = Bhaskara(1, -5, 6)
primeira_raiz = bhaskara.calcRaiz1()
segunda_raiz = bhaskara.calcRaiz2()

if primeira_raiz is None or segunda_raiz is None:
    print("Não existem raízes reais.")
else:
    print(f"Primeira raiz: {primeira_raiz}")
    print(f"Segunda raiz: {segunda_raiz}")
```
